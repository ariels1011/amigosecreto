#1. Análisis de facturación



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
# Nombre del archivo a analizar
file_name = 'tienda_1 .csv'
file_name = 'tienda_2 .csv'


try:
    # Cargar el archivo CSV
    df = pd.read_csv(url)
    df2 = pd.read_csv(url2)
    df3 = pd.read_csv(url3)
    df4 = pd.read_csv(url4)

    # Calcular la facturación (suma de la columna 'Precio')

    facturacion_tienda1 = round(df['Precio'].sum())
    facturacion_tienda2 = round(df2['Precio'].sum())
    facturacion_tienda3 = round(df3['Precio'].sum())
    facturacion_tienda4 = round(df4['Precio'].sum())

    # Formatear el resultado para mejor visualización
    facturacion_formato1 = f"${facturacion_tienda1:,}".replace(",", "X").replace(".", ",").replace("X", ".")
    facturacion_formato2 = f"${facturacion_tienda2:,}".replace(",", "X").replace(".", ",").replace("X", ".")
    facturacion_formato3 = f"${facturacion_tienda3:,}".replace(",", "X").replace(".", ",").replace("X", ".")
    facturacion_formato4 = f"${facturacion_tienda4:,}".replace(",", "X").replace(".", ",").replace("X", ".")

    print(f"La facturación total de tienda 1 es:")
    print(f"{facturacion_formato1}")
    print(f"La facturación total de tienda 2 es:")
    print(f"{facturacion_formato2}")
    print(f"La facturación total de tienda 3 es:")
    print(f"{facturacion_formato3}")
    print(f"La facturación total de tienda 4 es:")
    print(f"{facturacion_formato4}")

except FileNotFoundError:
    print(f"Error: El archivo {file_name} no fue encontrado.")
except KeyError:
    print(f"Error: La columna 'Precio' no se encontró en el archivo {file_name}. Por favor, verifica el nombre de la columna.")
except Exception as e:
    print(f"Ocurrió un error inesperado al procesar {file_name}: {e}")
```

# 2. Ventas por categoría Tienda 1

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)


tienda.head()
# Nombre del archivo a analizar
file_name = 'tienda_1 .csv'



try:
    # 1.Cargar el archivo CSV
    df = pd.read_csv(url)

    # 2. Agrupar por 'Categoría del Producto' y sumar la columna 'Precio'
    ventas_por_categoria = df.groupby('Categoría del Producto')['Precio'].sum()

    # 3. Ordenar el resultado de mayor a menor venta (opcional, pero útil)
    ventas_por_categoria = ventas_por_categoria.sort_values(ascending=False)

    # 4. Formatear y mostrar los resultados
    print("--- Ventas Totales por Categoría de Producto Tienda 1---")

    # Lista para almacenar los resultados formateados
    resultados_formateados = []

    for categoria, venta in ventas_por_categoria.items():
        # Convertir a entero (sin redondear)
        venta_entera = int(venta)

        # Formatear la venta con separadores de miles
        venta_formato = f"${venta_entera:,}".replace(",", "X").replace(".", ",").replace("X", ".")

        resultados_formateados.append(f"{categoria:<30}: {venta_formato}")

    for linea in resultados_formateados:
        print(linea)

    # Calcular y mostrar el total (sin decimales)
    total_facturacion = int(df['Precio'].sum())
    total_formato = f"${total_facturacion:,}".replace(",", "X").replace(".", ",").replace("X", ".")
    print("-" * 50)
    print(f"{'TOTAL DE FACTURACIÓN':<30}: {total_formato}")


except FileNotFoundError:
    print(f"Error: El archivo {file_name} no fue encontrado.")
except KeyError as e:
    print(f"Error: Una columna necesaria no fue encontrada. Verifique si 'Categoría del Producto' o 'Precio' existen. Detalle: {e}")
except Exception as e:
    print(f"Ocurrió un error inesperado al procesar {file_name}: {e}")
```

# 3. Calificación promedio de la tienda


```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)


tienda.head()
# Nombre del archivo a analizar
file_name = 'tienda_1 .csv'
file_name2 = 'tienda_2 .csv'
file_name3 = 'tienda_3 .csv'
file_name4 = 'tienda_4 .csv'



try:
    # 1.Cargar el archivo CSV
    df = pd.read_csv(url)
    df2 = pd.read_csv(url2)
    df3 = pd.read_csv(url3)
    df4 = pd.read_csv(url4)

    #
# 2. Calcular el promedio de la columna 'Calificación'<
    # Usamos .mean() para obtener el promedio
    promedio_calificacion = df['Calificación'].mean()
    promedio_calificacion2 = df2['Calificación'].mean()
    promedio_calificacion3 = df3['Calificación'].mean()
    promedio_calificacion4 = df4['Calificación'].mean()


    # 3. Formatear el resultado para mostrar dos decimales
    promedio_formato = f"{promedio_calificacion:.2f}"
    promedio_formato2 = f"{promedio_calificacion2:.2f}"
    promedio_formato3 = f"{promedio_calificacion3:.2f}"
    promedio_formato4 = f"{promedio_calificacion4:.2f}"

    print(f"--- Promedio de Calificación de Clientes ({file_name}) ---")
    print(f"El promedio de calificación de los clientes es: {promedio_formato}")
    print(f"--- Promedio de Calificación de Clientes ({file_name2}) ---")
    print(f"El promedio de calificación de los clientes es: {promedio_formato2}")
    print(f"--- Promedio de Calificación de Clientes ({file_name3}) ---")
    print(f"El promedio de calificación de los clientes es: {promedio_formato3}")
    print(f"--- Promedio de Calificación de Clientes ({file_name4}) ---")
    print(f"El promedio de calificación de los clientes es: {promedio_formato4}")

except FileNotFoundError:
    print(f"Error: El archivo {file_name} no fue encontrado.")
except KeyError:
    print(f"Error: La columna 'Calificación' no se encontró en el archivo {file_name}. Verifique el nombre de la columna.")
except Exception as e:
    print(f"Ocurrió un error inesperado al procesar {file_name}: {e}")
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)


tienda.head()
# Nombre del archivo a analizar
file_name = 'tienda_1 .csv'
file_name2 = 'tienda_2 .csv'
file_name3 = 'tienda_3 .csv'
file_name4 = 'tienda_4 .csv'



try:
    # 1.Cargar el archivo CSV
    df = pd.read_csv(url)
    df2 = pd.read_csv(url2)
    df3 = pd.read_csv(url3)
    df4 = pd.read_csv(url4)

# 2. Agrupar por 'Producto' y sumar la columna 'Cantidad' para obtener el total de unidades vendidas
    ventas_por_producto = df.groupby('Producto')['Cantidad de cuotas'].sum()

    # 3. Obtener los 5 productos más vendidos (top 5)
    productos_mas_vendidos = ventas_por_producto.nlargest(5)

    # 4. Obtener los 5 productos menos vendidos (bottom 5)
    productos_menos_vendidos = ventas_por_producto.nsmallest(5)

    # --- Mostrar Resultados ---
    print(f"--- 🥇 Top 5 Productos Más Vendidos ({file_name}) ---")
    for producto, cantidad in productos_mas_vendidos.items():
        print(f"{producto:<30}: {cantidad} unidades")

    print("\n" + "="*50 + "\n")

    print(f"--- 📉 Top 5 Productos Menos Vendidos ({file_name}) ---")
    for producto, cantidad in productos_menos_vendidos.items():
        print(f"{producto:<30}: {cantidad} unidades")

except FileNotFoundError:
    print(f"Error: El archivo {file_name} no fue encontrado.")
except KeyError as e:
    print(f"Error: Una columna necesaria no fue encontrada. Verifique si 'Producto' o 'Cantidad de cuotas' existen. Detalle: {e}")
except Exception as e:
    print(f"Ocurrió un error inesperado al procesar {file_name}: {e}")
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)


tienda.head()
# Nombre del archivo a analizar
file_name = 'tienda_1 .csv'
file_name2 = 'tienda_2 .csv'
file_name3 = 'tienda_3 .csv'
file_name4 = 'tienda_4 .csv'



try:
    # 1.Cargar el archivo CSV
    df = pd.read_csv(url)
    df2 = pd.read_csv(url2)
    df3 = pd.read_csv(url3)
    df4 = pd.read_csv(url4)
# 2. Calcular el promedio de la columna 'Costo de envío'
    promedio_costo_envio = df['Costo de envío'].mean()
    promedio_costo_envio2 = df2['Costo de envío'].mean()
    promedio_costo_envio3 = df3['Costo de envío'].mean()
    promedio_costo_envio4 = df4['Costo de envío'].mean()

    # 3. Formatear el resultado como moneda, sin decimales y con separadores de miles
    # Convertimos a entero para eliminar decimales antes de formatear
    costo_entero = int(promedio_costo_envio)
    costo_entero2 = int(promedio_costo_envio2)
    costo_entero3 = int(promedio_costo_envio3)
    costo_entero4 = int(promedio_costo_envio4)


    # Formato de moneda (punto para miles)
    costo_formato = f"${costo_entero:,}".replace(",", "X").replace(".", ",").replace("X", ".")
    costo_formato2 = f"${costo_entero2:,}".replace(",", "X").replace(".", ",").replace("X", ".")
    costo_formato3 = f"${costo_entero3:,}".replace(",", "X").replace(".", ",").replace("X", ".")
    costo_formato4 = f"${costo_entero4:,}".replace(",", "X").replace(".", ",").replace("X", ".")


    print(f"--- 🚚 Costo Promedio de Envío ({file_name}) ---")
    print(f"El costo promedio de envío por producto es: {costo_formato}")
    print(f"--- 🚚 Costo Promedio de Envío ({file_name2}) ---")
    print(f"El costo promedio de envío por producto es: {costo_formato2}")
    print(f"--- 🚚 Costo Promedio de Envío ({file_name3}) ---")
    print(f"El costo promedio de envío por producto es: {costo_formato3}")
    print(f"--- 🚚 Costo Promedio de Envío ({file_name4}) ---")
    print(f"El costo promedio de envío por producto es: {costo_formato4}")

except FileNotFoundError:
    print(f"Error: El archivo {file_name} no fue encontrado.")
except KeyError:
    print(f"Error: La columna 'Costo de envío' no se encontró en el archivo {file_name}. Verifique el nombre de la columna.")
except Exception as e:
    print(f"Ocurrió un error inesperado al procesar {file_name}: {e}")
```
